IMAGE TO PENCIL SKETCH USING PYTHON

TASK-2 Convert a image to a pencil sketch

```python
import numpy as np
import pandas as pd

import cv2 as cv
from google.colab.patches import cv2_imshow #image display
from skimage import io
from PIL import Image
import matplotlib.pyplot as plt
import urllib.request
```

```python
image= io.imread("rt.jpg")
cv2_imshow(image)
```

```python
grey_filter=cv.cvtColor(image,cv.COLOR_BGR2GRAY)
cv2_imshow(grey_filter)
```

```python
invert=cv.bitwise_not(grey_filter)
cv2_imshow(invert)
```

```python
blur=cv.GaussianBlur(invert,(21,21),0)
cv2_imshow(blur)
```

```python
invertedblur=cv.bitwise_not(blur)
cv2_imshow(invertedblur)
```

```python
sketch_filter=cv.divide(grey_filter,invertedblur,scale=255.0)
cv2_imshow(sketch_filter)
cv.imwrite("sketch.jpg",sketch_filter)
```
